# ML classification task for RWS and MHRW with sklearn random forest

```python
data = pd.read_csv('data/pca_ba_robust_example.csv')
```

```python
data.head()
```

```python
subpop_data = data[(data['method']=='RWS') | (data['method']=='MHRW')]
#subpop_data = subpop_data[subpop_data['len']==75]
```

```python
def method_map(row)->int:
    if row['method']=='RWS':
        return 1
    return 0
```

```python
subpop_data['indicator'] = subpop_data.apply(lambda row: method_map(row), axis=1)
```

```python
X = subpop_data.drop(['Component0', 'Component1', 'Component2', 'net id', 'method', 'len', 'indicator'], axis=1)
y = subpop_data['indicator']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import make_scorer, roc_auc_score
```

```python
def gini_score(expected, predicted, **unused_kwargs) -> float:
    return (roc_auc_score(expected, predicted)-0.5)*2.0
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=2112)
```

```python
#steps = [('scaler', StandardScaler()), ('model', RandomForestClassifier(n_estimators = 100, random_state = 42, n_jobs = -1))]
#model = Pipeline(steps = steps)
```

```python
model = RandomForestClassifier(n_estimators=200, max_depth=7, random_state=42, n_jobs=-1)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = np.round(model.predict_proba(X_test)[:, 1], 5)
```

```python
gini_score(y_test, y_pred)
```

```python
shap.initjs()
```

```python
Xshap = pd.concat([
    X_test[y_test==0], # IRWS
    X_test[y_test==1]  # TR
], axis=0).sample(frac=1)
```

```python
#explainer = shap.TreeExplainer(model[-1])
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(Xshap)
```

```python
shap_explanation = explainer(Xshap)
```

```python
shap.summary_plot(shap_values, Xshap, plot_size=(10,7), show=False)
plt.tight_layout()
plt.show()
```
